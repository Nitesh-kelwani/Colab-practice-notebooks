```python
# Count frequency of words in a strin
strr="This is the string containing the string this "
lst=strr.split(' ')
freq={}
for word in lst:
  freq[word]=freq.get(word,0)+1

print(freq)
```

```python
from collections import Counter
count=Counter(lst)
print(count)
```

```python
#  Linear Search
arr=[12,34,53,24,54,4,35]
target=35
j=0
found=False
for i in arr:
  if i==target:
    print(f"The target found at at  {j}th index")
    found=True
    break
  j+=1
if not found:
  print("Targtet not found")
```
