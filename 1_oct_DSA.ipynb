```python
# Rotate array (left/right)
# left rotate by 1
arr=[1,2,3,4,5]
left_rot=arr[1:]+arr[:1]
print(left_rot)
# right rotate by 1
arr=[1,2,3,4,5]
right_rot=arr[-1:]+arr[:-1]
print(right_rot)
```

```python
#  Palindrome for numbers and strings (with & without inbuilt)
#for number
num=121
n=num
r=0
while n>0:
  r,n=r*10+n%10,n//10
if num==r:
  print("It is a palindrome")
else:
  print("Not a palindrome")

```

```python
# for string
strr="nitin"
revstr=strr[::-1]
if strr==revstr:
  print("It is a palindrome")
else:
  print("Not a palindrome")
# builtin-function
rev=reversed(strr)
print("".join(rev))
rev=reversed(strr)
if strr=="".join(rev):
  print("It is a palindrome")
else:
  print("Not a palindrome")
```

```python
#  Explore Heat map & Implement that
import seaborn as sns
flights=sns.load_dataset('flights')
flights.head()
```

```python
x=flights.pivot_table(index="year",columns='month',values='passengers',aggfunc='sum')
sns.heatmap(x)
```

```python
def decorator(func):
    def wrapper(name):
        print(f"Hello {name}, welcome!")
        func(name)
        print("Goodbye!")
    return wrapper

@decorator
def greet(name):
    print(f"{name} is learning DS")

greet("clove")

```
