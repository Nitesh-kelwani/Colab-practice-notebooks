```python
# work with date module in python
from datetime import date,datetime,timedelta
```

```python
today=date.today()
print(today)
now=datetime.now()
print(now)
mybday=date(2004,11,14)
print(mybday)
future = today + timedelta(days=10)
print(future)
past = today - timedelta(days=5)
print(past)
```

```python
print(now.minute)
print(now.hour)
print(now.day)
print(now.month)
print(now.year)
```

```python
# formatted
formetted=now.strftime("%d-%m-%y %H-%M-%S")
print(formetted)
```
