```python
# Find missing number in array (1..n)
arr=[1,2,3,5]
n=5
sum_of_n_natural=n*(n+1)/2
print(sum_of_n_natural)
print(f"The missing value is :{int(sum_of_n_natural-sum(arr))}")
```

```python
#  Check Palindrome String
stri='nitin'
rever=stri[::-1]
if stri==rever:
  print("The string is palindrome")
else:
  print("String is not palindrome")
```

```python
# nested functions in python
def outer(name):
  def inner():
    print("hello", name)
  inner()

outer("Arlert")
```
