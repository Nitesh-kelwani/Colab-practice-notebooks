Reverse an array, Find min & max ( Sort in reverse order, no inbuilt functions)
sum individual digits in a number, draw a multiple line plots using figure and axes

```python
##reverse an arry
arr=[4,6,22,53,2,36,30]
rev=[]
length=len(arr)
for i in range(length):
  rev.append(arr[(length-1) - i])
print(rev)
```

```python
# Find min & max ( Sort in reverse order, no inbuilt functions)
lis=[4,6,22,53,2,36,30]
#using bubble sort for sorting the array in reverse

for i in range(len(lis)):
    swapped=False
    for j in range(0,len(lis)-i-1):
        if lis[j+1]>lis[j]:
            lis[j+1],lis[j]=lis[j],lis[j+1]
            swapped=True
    if swapped==False:
        break
print(lis)

print(f"Minimum is : {lis[len(lis)-1]}")
print(f"Minimum is : {lis[0]}")
```

```python
# sum individual digits in a number
num=78452
total=0
while num>0:
  digit=num%10
  total+=digit
  num=num//10

print(total)
```

```python
#  draw a multiple line plots using figure and axes
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[2,4,6,8,10]

plt.plot(x,y,label="Doubled")
plt.plot(x,[x**2 for x in x],label="squared")
plt.plot(x,[x**3 for x in x],label="cubed")
plt.show()
```
