```python
# Merge two sorted arrays (no inbuilt functions should be used)
arr1=[4,6,8,12,14]
arr2=[1,3,5,9]
len1=len(arr1)
len2=len(arr2)
j=0
i=0
sorted_arr=[]
while i<len1 and j<len2:
  if arr1[i]<=arr2[j] and i<=len1 :
    sorted_arr.append(arr1[i])
    i+=1
  else:
    sorted_arr.append(arr2[j])
    j+=1

while i<len1:
  sorted_arr.append(arr1[i])
  i+=1

while j<len2:
  sorted_arr.append(arr2[j])
  j+=1
print(sorted_arr)
```
