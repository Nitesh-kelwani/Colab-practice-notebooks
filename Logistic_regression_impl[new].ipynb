```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import make_classification
```

```python
#creating dataset
x,y=make_classification(n_samples=1000,n_features=10,n_classes=2,random_state=42)
```

```python
pd.DataFrame(x)
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.30,random_state=42)
```

```python
# Model training
from sklearn.linear_model import LogisticRegression
logistic=LogisticRegression()
```

```python
logistic.fit(xtrain,ytrain)
```

```python
ypred=logistic.predict(xtest)
print(ypred)
```

Performance metrics

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
score=accuracy_score(ytest,ypred)
print(score)
cm=confusion_matrix(ytest,ypred)
print(cm)
```

```python
print(classification_report(ytest,ypred))
```

Hyper Perameter Tuning and Cross Validation

```python
model=LogisticRegression()
penalty=['l1','l2','elasticnet']
c_values=[100,10,1.0,0.1,0.01]
solver=['lbfgs','liblinear','newton-cg','newton-cholesky','sag','saga']
```

```python
params=dict(penalty=penalty,C=c_values,solver=solver)
```

```python
from sklearn.model_selection import StratifiedKFold
cv=StratifiedKFold()
```

```python
##Grid Search CV
from sklearn.model_selection import GridSearchCV

grid=GridSearchCV(estimator=model,param_grid=params,scoring='accuracy',cv=cv,n_jobs=-1)
```

```python
grid
```

```python
grid.fit(xtrain,ytrain)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
ypred1=grid.predict(xtest)
```

```python
score=accuracy_score(ytest,ypred1)
print(score)
cm=confusion_matrix(ytest,ypred1)
print(cm)
print(classification_report(ytest,ypred1))
```

Randomized Search CV

```python
model1=LogisticRegression()
```

```python
from sklearn.model_selection import RandomizedSearchCV
randomcv=RandomizedSearchCV(estimator=model1,param_distributions=params,cv=5,scoring='accuracy')
```

```python
randomcv.fit(xtrain,ytrain)
```

```python
randomcv.best_score_
```

```python
randomcv.best_params_
```

```python
ypred1=randomcv.predict(xtest)
```

```python
score=accuracy_score(ytest,ypred1)
print(score)
cm=confusion_matrix(ytest,ypred1)
print(cm)
print(classification_report(ytest,ypred1))
```
